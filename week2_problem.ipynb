<h3>연속 신호와 이산 신호</h3>
<p><b>신호의 길이(지속 시간(구간))</b>
<ul> 
    <li>연속 신호: 신호 값이 존재하는 시간 구간의 길이 <i>L</i>로 정의
    <font color="blue">x(t)가 t1<=t<=t2 에서 정의 -> <i><b>L=t2-t1</b></i></font></li> 
    <li>이산 신호: 신호를 이루는 샘플링의 개수 <i>N</i>로 정의
    <font color="blue">x[n]이 n1<=n<=n2 에서 정의 -> <i><b>N=n2-n1+1</b></i></font></li>
</ul>
<p><b>유한 구간 신호와 무한 구간 신호</b></p>
<ul>
    <li>유한 구간 신호: 시간적으로 유한한 길이를 갖는 신호</li>
    <li>무한 구간 신호: 시간적으로 무한한 길이를 갖는 신호</li>
</ul>
<p><b>연속 신호의 샘플링에 의한 이산 신호
<hr>

<b>예제 1-3</b>, 다음의 신호를 그리고, 신호의 길이를 구하라.

(a). x(t)=sin(0.5\*pi\*t)(if -1<=t<=3), 0(else)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, pi

def x(t):
    if -1 <= t <= 3:
        return sin(0.5 * pi * t)
    else:
        return 0

#이산 신호를 샘플링 할 시간 간격 설정(1/3초)
sampling_interval=1/3

# t 값을 생성합니다.
t_values = np.linspace(-2, 4, 400)  # -2부터 4까지 400개의 점 생성
t_values_sampling=np.arange(-1, 3+sampling_interval, sampling_interval)

# x(t) 값을 계산합니다.
x_values = [x(t) for t in t_values]
x_values_sampling=[x(t) for t in t_values_sampling]

# 그래프를: 연속 신호
plt.plot(t_values, x_values)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('''a's continuous signal''')
plt.grid(True)
plt.show()

#그래프_이산 신호
plt.stem(t_values_sampling, x_values_sampling, use_line_collection=True)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('''a's discrete signal''')
plt.grid(True)
plt.show()
```

<b>b. x(t)=cos(2\*pi\*t)_(if 0\<=t\<=5), 0_(else)</b>

```python
import numpy
from math import cos, pi
import matplotlib.pyplot as plt

def x(t):
    if t>=0 and t<=5:
        return cos(2*pi*t)
    else:
        return 0
    
#변수 지정_연속 신호
t_values2=np.linspace(-1, 6, 400)
x_values2=[x(t) for t in t_values2]

#그래프_연속 신호
plt.plot(t_values2, x_values2)
plt.xlabel=('t')
plt.ylabel=('x(t)')
plt.title("b's continous signal")
plt.grid(True)
plt.show()

#변수 지정_이산 신호
sampling_interval2=0.5
t_values2_samp=np.arange(0, 5+sampling_interval2, sampling_interval2)
x_values2_samp=[x(t) for t in t_values2_samp]

#그래프_이산 신호
plt.stem(t_values2_samp, x_values2_samp, use_line_collection=True)
plt.xlabel('t')
plt.ylabel("x(t)")
plt.title('''b's discrete signal''')
plt.grid(True)
plt.show
```
