### 2.2.1 연속 신호와 이산 신호
* 연속 신호와 이산 신호
    * 연속(시간)신호 $x(t)$: 시간에 대해 끊어지지않는,즉 모든 ㅣ간에 대해 정의되는 신호
    * 이산(시간) 신호 $x[n]$: 띄엄띄엄 특정한 시각에서만 정의되는 신호

* 아날로그 신호와 디지털 신호
    - __시간 뿐만 아니라 크기(값)에 대해서도 연속이냐 이산이냐 구분__
    * 아날로그 신호: 시간과 크기에 대해 모두 연속인 신호
        * 예) ECG, EEG

    * 디지털 신호: 시간과 크기에 대해 모두 이산인 신호
        * 예) 컴퓨터의 이진 신호

#### 아날로그/디지털(Analog-to-Digital: A/D)
**- 연속 신호를 디지털 신호로 바꾸는 동작**
**- 샘플링+양자화+부호화**

* 샘플링: 아날로그 신호를 이산 신호로 변환
    * 일정한 시간 간격으로 연속 신호의 순시값을 취하는 동작
* 양자화: 이산 신호를 디지털 신호로 변환
    * 크기에 대해서도 이산화시켜 신호의 값을 유한개의 등급으로 분류
* 부호화: 디지털 신호를 이진 부호로 변환 ($\in{0, 1}$)
    * 디지털 시스템이 받아들여 처리할 수 있는 형태의 부호로 바꿈

#### 주기 신호와 비주기 신호
* 주기 신호
    * 동일한 파형이 일정 시간마다 끊임없이 반복되는 신호
       $x[n+kN]=x[n], k, N은 정수$   
       예: 정현파, 컴퓨터의 clock pulse
* 비주기 신호
    * 주기성을 만족하지 않는 신호
*기본 주기
    * 파형이 반복되는 (최소)시간 간격 $T(또는 N)$

* 두 주기 신호의 합이 주기 신호가 될 조건
    * 각 주기 신호의 주기의 비가 유리수일 경우에 한해 주기 신호가 됨

    $\frac{T_1}{T_2}(또는 \frac{N_1}{N_2})=\frac{l}{k}$

### 예제 2-1. 주기 신호의 판별
다음 신호가 주기 신호인지 판별하고, 만약 주기 신호라면 주기를 구하라

(a). $x(t)=cos(\frac{\pi}{3}t)+sin(\frac{\pi}{2}t)$

(b). $y(t)=cos(\frac{1}{3}t)+sin(\frac{1}{2}t)$

(c). $z(t)=cos(\frac{\pi}{3}t)+sin(\frac{1}{2}t)$

(d). $x[n]=sin(\frac{3\pi}{4}n)+cos(\frac{\pi}{3}n)$

````python
#a
import numpy as np
import matplotlib.pyplot as plt
from math import cos, sin, pi, gcd

def xa(t):
    return cos((pi/3)*t)+sin((pi/2)*t)
def xa1(t):
    return cos((pi/3)*t)
def xa2(t):
    return sin((pi/2)*t)
Aomega1=pi/3
#(2*pi)/T1=pi/3, T1=(2*pi*3)/pi
T1a=(2*pi*3)/pi

Aomega2=1/2
#(2*pi)/T2=pi/2, T2=(2*pi*2)/pi
T2a=(2*pi*2)/pi

g=gcd(int(T1a), int(T2a))
Ta=int((T1a*T2a)/g)

print(f"주기 T1: {T1a}, 주기 T2: {T2a}, 최소공배수를 통한 주기 T: {Ta}")

t_value=np.arange(0, 20, 0.01)
x_value_total=[xa(t) for t in t_value]
x_value1=[xa1(t) for t in t_value]
x_value2=[xa2(t) for t in t_value]

plt.plot(t_value, x_value_total, c="red")
plt.plot(t_value, x_value1, c="blue", linestyle="--")
plt.plot(t_value, x_value2, c="green", linestyle="--")
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
````

````python
#b
def yb(t):
    return cos((1/3)*t)+sin((1/2)*t)
def yb1(t):
    return cos((1/3)*t)
def yb2(t):
    return sin((1/2)*t)
Bomega1=1/3
T1b=(2*pi)/Bomega1

Bomega2=1/2
T2b=(2*pi)/Bomega2

g=gcd(int(T1b), int(T2b))
period_b=(T1b*T2b)/g
print(f"T1: {T1b/pi}, T2: {T2b/pi}, T: {int(period_b/pi)}")

t_value_b=np.arange(0, 50, 0.01)
y_value1=[yb1(t) for t in t_value_b]
y_value2=[yb2(t) for t in t_value_b]
y_value=[yb(t) for t in t_value_b]

plt.plot(t_value_b, y_value1, linestyle='--' , c='blue')
plt.plot(t_value_b, y_value2, linestyle='--', c='green')
plt.plot(t_value_b, y_value, c='red')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True)
plt.show()
````

````python
#c
def zc(t):
    return cos((pi/3)*t)+sin((1/2)*t)
def zc1(t):
    return cos((pi/3)*t)
def zc2(t):
    return sin((1/2)*t)

Comega1=pi/3
T1c=(2*pi)/Comega1

Comega2=1/2
T2c=(2*pi)/Comega2
g=gcd(int(T1c), int(T2c))
period_c=(T1c*T2c)/g

if(T1c//T2c==int):
    print("주기적 O")
    print(f"T1: {T1c}, T2: {T2c}, T: {period_c}")
else:
    print("주기적 X")

t_value_c=np.arange(0, 50, 0.01)
z_value1=[zc1(t) for t in t_value_c]
z_value2=[zc2(t) for t in t_value_c]
z_value=[zc(t) for t in t_value_c]

plt.plot(t_value_c, z_value1, linestyle="--", c="blue")
plt.plot(t_value_c, z_value2, linestyle="--", c="green")
plt.plot(t_value_c, z_value, c="red")
plt.xlabel('t')
plt.ylabel('z(t)')
plt.grid(True)
plt.show()
````

#### 2.1.3 우대칭 신호와 기대칭 신호
**우대칭 신호와 기대칭 신호**
* 우대칭 신호: 파형이 세로축에 대해 대칭인 신호
   * 예) 2차함수, 코사인함수 등
   
$x(t)=x(-t)$
* 기대칭 신호: 파형이 원첨에 대해 $180\degree$대칭인 신호
   * 예) 사인함수, 탄젠트함수, 일차함수 등
   
   $x(t)=-x(-t)$
* 모든 임의의 신호는 우대칭 신호와 기대칭 신호의 합으로 표현 가능

   $x(t)=\frac{x(t)+x(-t)}{2}+\frac{x(t)-x(-t)}{2}=x_e(t)+x_o(t)$

#### 예제 2-2. 신호의 우대칭 성분과 기대칭 성분
$x(t)=cos(2\pi t)+sin(4\pi t)$의 우대칭 성분과 기대칭 성분을 구하라

````python
def x(t):
    return cos(2*pi*t)+sin(4*pi*t)
def x1(t):
    return cos(2*pi*t)
def x2(t):
    return sin(4*pi*t)

def xe(t):
    return (x(t)+x(-t))/2

t_value=np.arange(0, 5, 0.01)
x1_value=[x1(t) for t in t_value]
x2_value=[x2(t) for t in t_value]
x_value=[x(t) for t in t_value]
xe_value=[xe(t) for t in t_value]

plt.plot(t_value, x1_value, c='blue', linestyle='--')
plt.plot(t_value, x2_value, c='green', linestyle='--')
plt.plot(t_value, x_value, c='yellow', linestyle='--')
plt.plot(t_value, xe_value, c='red')
plt.xlabel('t')
plt.ylabel('x')
plt.grid(True)
plt.show()
````
