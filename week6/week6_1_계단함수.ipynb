-> 불연속적인 파형의 구간 연속 신호를 표현하는 데 매우 유용

### 예제 3-1. 계단 함수로 표현된 신호의 파형

다음에 주어진 신호를 그려라

(a). $x(t)=u(t+2)-2u(t)+u(t-1)$

(b). $x(t)=t[u(t+1)-u(t-2)]$

```python
#a번 문제
#x(t)=[u(t+2)-u(t)]+(-1)[u(t)-u(t-1)]
def xa(t):
    if (t>-2 and t<0): return 1
    elif (t>0 and t<1): return -1
    else: return 0

t_value=np.linspace(-3, 3, 400)
xa_value=[xa(t) for t in t_value]
plt.plot(t_value, xa_value)
plt.show()
```

```python
#(b)번문제
#xb(t)=t[u(t+1)-u(t-2)]

#계단함수 u(t) 정의
def u(t):
    return np.where(t>=0, 1, 0)

#신호xb(t) 정의
def xb(t):
    return t*(u(t+1)-u(t-2))

t=np.linspace(-2, 3,1000)
x_values=xb(t)

plt.plot(t, x_values)
plt.xlabel('t')
plt.ylabel('xb(t)')
plt.grid(True)
plt.show()
```
