> * 디랙(Dirac) 델타: $t=0$ 순간에만 존재 & 면적이 1이고 우대칭 함수

$\left\{\begin{matrix}
1\space \delta(t) \to \infty, (t=0) \\
2\space \delta(t)=0, (t\neq0)\\
3\space \int_{\infty}^{\infty}\delta(t)=1\\
4\space \delta(t)=\delta(-t)\\
\end{matrix}\right.$

![Alt text](image-4.png)

$\left\{\begin{matrix}
\delta(t)=\frac{du(t)}{dt}\\
u(t)=\int_{-\infty}^{t}\delta(\tau)d\tau\\
\end{matrix}\right.$

> * 임펄스 함수의 체 거르기(shifting) 성질(샘플링 성질)

$x(t)=\delta(t-t_0)=x(t_0)delta(t-t_0)$

$\int_{-\infty}^{\infty}x(t)\delta(t-t_0)dt=x(t_0)\int_{-\infty}^{\infty}\delta(t')dt'=x(t_0)$

$=\int_{-\infty}^{\infty}x(t)\delta(t-t_0)dt=x(t_0)\int_{-\infty}^{\infty}\delta(t-t_0)dt$
