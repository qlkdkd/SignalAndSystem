```python
import numpy as np
import matplotlib.pyplot as plt
```

### 3.1.3. 부호 함수
$sgn=\left\{\begin{matrix}
-1,(t<0) \\
0, (t=0)\\
+1, (t>0)\end{matrix}\right.$

```python
def sgn(t):
    if(t<0): return -1
    elif (t==0): return 0
    else: return 1

t_values=np.linspace(-1, 1, 1000)
x_values=[sgn(t) for t in t_values]
plt.plot(t_values, x_values)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

### 3.1.4. 램프 함수
$r(t)=\left\{\begin{matrix}
t,(t>=0) \\
0, (t<0)\end{matrix}\right.$

$\int_{-\infty}^{t}u(\tau)d\tau=tu(t)=r(t)$
![](image-2.png)

* 계단 함수와 램프함수: 시스템 특성 파악을 위한 시험 신호로 많이 활용
    * 계단 신호: 급작스런 입력 신호 값의 변화에 대한 시스템 반응
    * 램프 신호: 시간에 비례하여 증가하는 입력 신호에 대한 시스템 반응

```python
def r(t):
    if(t>=0): return t
    else: return 0

t_value=np.linspace(-1, 1, 1000)
x_value=[r(t) for t in t_value]
plt.plot(t_value, x_value)
plt.xlabel('t')
plt.ylabel('r(t)')
plt.grid(True)
plt.show()
```

### 3.1.5. 삼각 펄스 함수

$tri(t)=\left\{\begin{matrix}
1-|t|, (t<=1)\\
0, (|t|>1)\end{matrix}\right.$

$tri(t)=(t+1)[u(t+1)-u(t)]+(-t+1)[u(t)-u(t-1)]$

$=(t+1)u(t+1)-2tu(t)+(t-1)u(t-1)$

$r(t+1)-2r(t)+r(t-1)$

![Alt text](image-3.png)

```python
def tri(t):
    if(abs(t)<=1): return 1-abs(t)
    elif(abs(t)>1): return 0

t_value=np.linspace(-2, 2, 1000)
x_value=[tri(t) for t in t_value]
plt.plot(t_value, x_value)
plt.xlabel('t')
plt.ylabel('tri(t)')
plt.grid(True)
plt.show()
```
