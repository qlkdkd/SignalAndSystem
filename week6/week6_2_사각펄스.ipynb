### 예제 3-2. 사각 펄스(계단함수)를 이용한 구간 연속 신호의 표현
그림 3-6에 나타낸 신호 $x(t)$의 수식 표현을 구하라

```python
def x(t):
    if (t>=-2 and t<=-1): return t+2
    elif (t>=-1 and t<=0): return 1
    elif (t>0 and t<2): return -1
    else: return 0

t_values=np.linspace(-3, 3,1000)
x_values=[x(t) for t in t_values]
plt.plot(t_values, x_values)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

$x(t)=(t+2)[u(t+2)-u(t+1)]+[u(t+1)-u(t)]+(-1)[u(t)-u(t-2)]$

$= (t+2)u(t+2)-(t+1)u(t+1)-2u(t)+u(t-2)$
