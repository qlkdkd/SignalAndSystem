```python
import numpy as np
import matplotlib.pyplot as plt

A = 5  # 진폭
T = 2  # 주기
tau = 0.2  # tau 값
k_values = np.arange(0, 30)  # k 값 범위

def Xk(k):
    return np.sinc(k/5)

Xk_values=np.array([Xk(k) for k in k_values])
spectrum=np.abs(Xk_values)
angle=np.angle(Xk_values)

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.stem(k_values, spectrum, use_line_collection=True)
plt.title('amplitude spectrum')
plt.xlabel('k')
plt.ylabel('Re(Xk)')

plt.subplot(122)
plt.stem(k_values, angle, use_line_collection=True)
plt.title('phase spectrum')
plt.xlabel('k')
plt.ylabel('|Xk|')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1  # 진폭
T = 2  # 주기
tau = 0.5  # tau 값

# 주파수 계산
omega_0 = 2 * np.pi / T

# 스펙트럼 계수 X_k 계산
def X_k(k):
    return np.sinc(k / 2)

# 시간 t 범위 설정
t = np.linspace(-10, 10, 1000)

# 조파 합성 x(t) 계산
def x(t):
    sum_value = 0
    for k in range(-10, 11):  # 유한한 범위로 합성
        sum_value += X_k(k) * np.exp(1j * k * omega_0 * t)
    return A * sum_value

# x(t) 계산
x_values = np.real(x(t))

# 그래프 그리기
plt.plot(t, x_values)
plt.title('주어진 조파 합성 x(t)')
plt.xlabel('시간 (t)')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```
