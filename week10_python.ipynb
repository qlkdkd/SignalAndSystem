```python
import numpy as np
import matplotlib.pyplot as plt

A = 5  # 
T = 2  # 
tau = 0.2  # tau 
k_values = np.arange(0, 30)  # k  

def Xk(k):
    return np.sinc(k/5)

Xk_values=np.array([Xk(k) for k in k_values])
spectrum=np.abs(Xk_values)
angle=np.angle(Xk_values)

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.stem(k_values, spectrum, use_line_collection=True)
plt.title('amplitude spectrum')
plt.xlabel('k')
plt.ylabel('Re(Xk)')

plt.subplot(122)
plt.stem(k_values, angle, use_line_collection=True)
plt.title('phase spectrum')
plt.xlabel('k')
plt.ylabel('|Xk|')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

A = 1  # 
T = 2  # 
tau = 0.5  # tau 

#  
omega_0 = 2 * np.pi / T

#   X_k 
def X_k(k):
    return np.sinc(k / 2)

#  t  
t = np.linspace(-10, 10, 1000)

#   x(t) 
def x(t):
    sum_value = 0
    for k in range(-10, 11):  #   
        sum_value += X_k(k) * np.exp(1j * k * omega_0 * t)
    return A * sum_value

# x(t) 
x_values = np.real(x(t))

#  
plt.plot(t, x_values)
plt.title('   x(t)')
plt.xlabel(' (t)')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```
