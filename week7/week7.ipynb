# 3.2. 연속 신호에 대한 기본 연산
## 3.2.1. 두 신호의 합과 곱
$$y(t)=x(t)+s(t)$$
$$y(t)=x(t)\cdot s(t)$$

### 예제 3.6. 신호의 합과 곱
다음 신호의 파형을 그려라

(a). $x(t)=e^{-t}cos(2\pi t)u(t)$

(b). $y(t)=u(t)-x(t)$

```python
import numpy as np
import matplotlib.pyplot as plt

def e(t):
    return np.exp(-t)

def c(t):
    return np.cos(2*np.pi*t)

def u(t):
    if t>0: return 1
    else: return 0

def x(t):
    return e(t)*c(t)*u(t)

def y(t):
    return u(t)-x(t)

t_val=np.linspace(0, 5, 400)
x_val=[x(t) for t in t_val]
y_val=[y(t) for t in t_val]
# a번 그래프
plt.plot(t_val, x_val)
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('$x(t)=e^{-t}cos(2\pi t)u(t)$')
plt.ylim(-2, 2)
plt.grid(True)
plt.show()
#b번 그래프
plt.plot(t_val, y_val)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('$y(t)=u(t)-x(t)$')
plt.grid(True)
plt.show()
```

## 3.2.2 신호의 진폭 변환
### 진폭 반전
* 시간축에 대해 대칭: 값의 부호가 바뀜
$$y(t)=-x(t)$$
$$=-y(t)=x(t)$$
> $-y=f(x), y=f(-x)$(y축 대칭)
* 진폭 이동: 파형의 변화 없이 세로축을 따라 평행 이동
$$y(t)=x(t)+a$$


### 진폭 척도 조절: 증폭(a>1)& 감쇠(a<1)
* 진폭의 값을 일정한 비율로 바꾸는 것
$$y(t)=ax(t)$$

### 예제 3.7. 신호의 진폭 변환
[그림 3.18(a)]의 신호 x(t)에 대해 다음의 신호를 그려라

a. $s(t)=2x(t)$
b. $v(t)=-s(t)$
c. $z(t)=v(t)+2$

```python
def x(t):
    # T를 임의로 1로 설정함
    if (t>0 and t<=1): return t
    elif (t>1 and t<2): return 1
    else: return 0
    
#(a)번 문제
def s(t):
    return 2*x(t)
#(b)번 문제
def v(t):
    return -s(t)
#(c)번 문제
def z(t):
    return v(t)+2
t_val=np.linspace(0, 3, 400)
s_val=[s(t) for t in t_val]
v_val=[v(t) for t in t_val]
z_val=[z(t) for t in t_val]

#a
plt.plot(t_val, s_val)
plt.xlabel('t')
plt.ylabel('s(t)')
plt.show()
#b
plt.plot(t_val, v_val)
plt.ylabel('v(t)')
plt.show()
#c
plt.plot(t_val, z_val)
plt.ylabel('z(t)')
plt.show()
```

## 3.2.3. 신호의 시간 변환
### 시간 반전(반사)
* 시간축을 거꾸로 뒤집는 것(t -> -t)
$$y(t)=x(-t)$$
*예: 비디오/오디오 테이프를 거꾸로 재생

### 시간 이동: 시간 지연(t_0>0)&시간 앞섬(t_0<0)
* 파형의 변화 없이 신호를 시간에 대해 이동
$$y(t)=x(t-t_0)$$
* 예: 번개와 천둥 소리, TV에서 특파원 호출 현장 중계


### 시간 척도 조절: 늘이기(신장)(a<1)&줄이기(압축)(a>1)
* 시간축에 대해 늘이기 또는 줄이기를 하여 척도를 바꾸는 동작
$$y(t)=x(at)$$
* 예: TV의 슬로우 모션, 인터뷰 음성 변조


### 예제 3.8. 신호의 시간 변환
그림[3-21(a0]의 신호 $x(t)$에 대해 다음의 신호를 그려라

$a. s(t)=x(-t)$
$b. v(t)=s(t-4)$
$c. z(t)=v(2t)$

```python
def x(t):
    if(t>0 and t<=2): return 0.5*t
    elif (t>1 and t<4): return 1
    else: return 0
def s(t):
    return x(-t)
def v(t):
    return s(t-4)
def z(t):
    return v(2*t)

t_val=np.linspace(-5, 5, 400)
x_val=[x(t) for t in t_val]
s_val=[s(t) for t in t_val]
v_val=[v(t) for t in t_val]
z_val=[z(t) for t in t_val]

plt.plot(t_val, x_val)
plt.title('x(t)')
plt.show()
plt.plot(t_val, s_val)
plt.title('s(t)')
plt.show()
plt.plot(t_val, v_val)
plt.title('v(t)')
plt.show()
plt.plot(t_val, z_val)
plt.title('z(t)')
plt.show()
```

## 3.2.5. 연산 조합에 의한 신호의 표현
### 연산의 조합
* 복잡한 신호도 단순한 신호에 기본 연산의 조합을 수행해 얻을 수 있음

예) $$y(t)=-2x(-\frac{t}{2}+1)+1$$

    * 시간반전, 시간 척도 조절, 시간이동
    * 진폭반전, 진폭 척도 조절, 진폭이동
* 일반적으로 연산의 적용 순서가 뒤바뀌더라도 같은 결과를 얻음
> **연산 순서를 바꾸면 시간 이동 값 달라짐!**

### 예제 3.9. 연속 신호에 대한 기본 연산의 조합
[그림 3-21(a)]의 신호 $x(t)$에 대해 $y(t)=x(-2t+4)$를 구하여 그려라.

```python
# y(t)=x(2(-t-2)) (시간에 대해 척도조절-반전-이동)
# 1. 시간 척도 조절: x_1(t)=x(2t)
# 2. 시간 반전: x_2(t)=x_1(-t)
# 3. 시간 이동: x_3(t)=x_2(t-2)

# y(t)=x(2(-t)+4) (시간에 대해 이동-반전-척도조절)
# 1. 시간 이동: x_4t(t)=x(t+4)
# 2. 시간 반전: x_5(t)=x_4(-t)
# 3. 시간 척도 조절: x_6(t)=x_5(2t)
def y(t):# x(t)는 예제 3.8에 정의되어있음
    return x(-2*t+4)

t_val=np.linspace(-4, 4, 1000)
y_val=[y(t) for t in t_val]

plt.plot(t_val, y_val)
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```

### 예제 3.10. 연속 신호에 대한 기본 연산의 조합
[예제 3-1]의 그림 3-3(a)의 신호 $x(t)$에 대해 $y(t)=-2x(-\frac{t}{2}+1)$을 구하여 그려라.

```python
def u(t):
    if(t>0): return 1
    else: return 0
def x(t):
    return np.exp(-t)*np.cos(2*np.pi*t)*u(t)
def y(t):
    return -2*x((-t/2)+1)

t_val=np.linspace(-5, 5, 1000)
y_val=[y(t) for t in t_val]
plt.plot(t_val, y_val)
plt.xlabel('t'); plt.ylabel('y(t)'); 
plt.show()
```
